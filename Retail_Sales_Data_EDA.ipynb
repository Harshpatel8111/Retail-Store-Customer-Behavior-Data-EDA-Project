```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
df['TransactionDate'].head()
```

Transformated Transaction Date to DateTime 

```python
df['Month'] = df['TransactionDate'].dt.month
df['Day'] = df['TransactionDate'].dt.day
df['Year'] = df['TransactionDate'].dt.year
```

```python
df.head()
```

```python
DateString= df['TransactionDate'].to_string()
```

```python
split = DateString.split(' ')
```

```python
df['TransDate'] = split[0]
df['TransTime'] = split[1]
```

```python
df.head()
```

```python
DateString
```

```python
df.drop('TransDate',axis=1,inplace=True)
df.drop('TransTime', axis = 1, inplace=True)
```

```python
df.head()
```

```python
df['TransactionDate'].head()
```

```python
df['TransactionDate'] = df['TransactionDate'].astype(str)
df['TransDate'] = df['TransactionDate'].str.split(" ").str[0]
df['TransTIme'] = df['TransactionDate'].str.split(" ").str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
# Convert Returned to int
df['Returned'] = df['Returned'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
df['TransTime'] = df['TransTIme']
```

```python
df['TransHour'] = df['TransTime'].str.split(':').str[0]
```

```python
df.drop(columns=['TransactionDate','TransTIme','TransTime'], inplace= True)
```
